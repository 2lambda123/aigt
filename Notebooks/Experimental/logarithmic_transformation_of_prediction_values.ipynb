```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [(1.0/2**x) for x in range(20)]
zzz = plt.plot(a)
```

```python
e = 3
b = np.clip(a, 10**(-e), 1.0)
c = np.log10(b*(10**e))/e

zzz = plt.plot(c)
```

```python
e = 4
c = np.log10(np.clip(a, 10**(-e), 1.0)*(10**e))/e

zzz = plt.plot(c)
```
