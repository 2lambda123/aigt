```python
import os
import matplotlib.pyplot as plt
import numpy as np
from local_vars import root_folder
import random 
```

```python
num_volume_slices = 3
num_samples = 3
```

```python
image_folder = r"\Ultrasound\NeedleCapture\2020-03-16-NeedleCapture-Medium\3Slice\Results"
```

```python
idxs = random.sample(range(1, 200), num_samples) 
```

```python
# plot source, generated and target images
def plot_images(src_img, gen_img, tar_img, comb_img):
    images = np.dstack((src_img, gen_img, tar_img, comb_img))
    # scale from [-1,1] to [0,1]
    images = (images + 1) / 2.0
    titles = ['Generated', 'Expected', 'Combined']
    # plot images row by row
    for i in range(num_volume_slices+3):
        # define subplot
        plt.subplot(1, 8, 1 + i)
        # turn off axis
        plt.axis('off')
        
        # plot raw pixel data
        if i < num_volume_slices:
            plt.imshow(images[:,:,i])
            plt.title('Source')
        else:
            plt.imshow(images[:,:,i])
            plt.title(titles[i-(num_volume_slices)])
        
    plt.show()
```

```python
for idx in idxs:
    segmentation_file = "NeedleCapture_Tofu_02_{:03d}_source.npy".format(idx)
    ultrasound_file = "NeedleCapture_Tofu_02_{:03d}_target.npy".format(idx)
    generated_file = "NeedleCapture_Tofu_02_{:03d}_generated.npy".format(idx)
    combined_file = "NeedleCapture_Tofu_02_{:03d}_combined.npy".format(idx)
    
    segmentation_np = np.load( os.path.join(root_folder, image_folder, segmentation_file) )
    ultrasound_np = np.load( os.path.join(root_folder, image_folder, ultrasound_file) )
    generated_np = np.load( os.path.join(root_folder, image_folder, generated_file) )
    combined_np = np.load( os.path.join(root_folder, image_folder, combined_file) )
    print(idx)
    plt.rcParams["figure.figsize"]=20,20
    plot_images(segmentation_np, generated_np, ultrasound_np, combined_np)
```

```python
segmentation_file = r"\Ultrasound\NeedleCapture\2020-03-16-NeedleCapture-Medium\3Slice\Results\NeedleCapture_Tofu_02_001_source.npy"
ultrasound_file = r"\Ultrasound\NeedleCapture\2020-03-16-NeedleCapture-Medium\3Slice\Results\NeedleCapture_Tofu_02_001_target.npy"
generated_file = r"\Ultrasound\NeedleCapture\2020-03-16-NeedleCapture-Medium\3Slice\Results\NeedleCapture_Tofu_02_001_generated.npy"
combined_file = r"\Ultrasound\NeedleCapture\2020-03-16-NeedleCapture-Medium\3Slice\Results\NeedleCapture_Tofu_02_001_combined.npy"
```

```python
segmentation_np = np.load( os.path.join(root_folder, segmentation_file) )
ultrasound_np = np.load( os.path.join(root_folder, ultrasound_file) )
generated_np = np.load( os.path.join(root_folder, generated_file) )
combined_np = np.load( os.path.join(root_folder, combined_file) )
```

```python
plot_images(segmentation_np, generated_np, ultrasound_np, combined_np)
```
